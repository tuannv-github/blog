# Subplots

### Returns
* ```fig``` Figure. The top level container for all the plot elements.
* ```ax```  Axes or array of Axes. An Axes object encapsulates all the elements of an individual (sub-)plot in a figure.

``` python
        # using the variable ax for single a Axes
        fig, ax = plt.subplots()

        # using the variable axs for multiple Axes
        fig, axs = plt.subplots(2, 2)

        # using tuple unpacking for multiple Axes
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
```

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
````

````python
# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.set_title('Simple plot')
ax.plot(x, y)
````

````python
# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.set_title('Plot')
ax2.set_title('Scatter')
ax1.plot(x, y)
ax2.scatter(x, y)
````

````python
# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)
````

````python
# Share a X axis with each column of subplots
fig, axs = plt.subplots(2, 2, sharex='col')
fig.suptitle("Share a X axis with each column of subplots")

# Share a Y axis with each row of subplots
fig, axs = plt.subplots(2, 2, sharey='row')
fig.suptitle("Share a Y axis with each row of subplots")

# Share both X and Y axes with all subplots
fig, axs = plt.subplots(2, 2, sharex='all', sharey='all')
fig.suptitle("Share both X and Y axes with all subplots")

# Note that this is the same as
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
fig.suptitle("Share both X and Y axes with all subplots")

# Create figure number 10 with a single subplot
# and clears it if it already exists.
fig, ax = plt.subplots(num=10, clear=True)
fig.suptitle("Figure number 10 with a single subplot")
````
